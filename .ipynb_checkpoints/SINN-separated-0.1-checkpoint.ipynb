```python
def noise_sample(obs_batch):
    obs_batch = obs_batch.cpu().detach().numpy()
    sample = np.random.normal(size=obs_batch.shape)
    sample = torch.from_numpy(sample).float().cuda()# use .to(self.device) soon 
    return sample

def noise_sample_step(obs_batch, epsilon=1):
    obs_batch = obs_batch.cpu().detach().numpy()
    step = np.random.normal(size=obs_batch.shape)
    step = (step / np.linalg.norm(step)) * epsilon
    z_obs_batch = obs_batch + step
    z_obs_batch = torch.from_numpy(z_obs_batch).float().cuda()# use .to(self.device) soon 
    return z_obs_batch

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(784, 800)

    def forward(self, x):
        x = self.fc1(x)
        return x
    
from scipy.linalg import svd

def orth(A):
    u, s, vh = svd(A, full_matrices=False)
    return u
```

```python
class RNDModel(nn.Module):
    def __init__(self, n_classes, dim=784):
        super(RNDModel, self).__init__()
        
        self.activated_target = None
        self.activated_predictor = None
        self.activated_predictor_next = None
        self.targets = {}
        self.predictors = {}
        self.predictors_next = {}
        
        self.optimizers = {}
        self.optimizers_next = {}

        for c in range(n_classes):
            self.targets[f'{c}'] = Net()             
            self.predictors[f'{c}'] =  Net()
            self.predictors_next[f'{c}'] = Net()
            
            self.optimizers[f'{c}'] = \
            optim.Adam(self.predictors[f'{c}'].parameters(),
                       0.00001)
            
            self.optimizers_next[f'{c}'] = \
            optim.Adam(self.predictors_next[f'{c}'].parameters(),
                       0.00001)
        

        for p in self.modules():
            if isinstance(p, nn.Linear):
                init.orthogonal_(p.weight)
        
        for target in self.targets:
            for param in self.targets[target].parameters():
                param.requires_grad = False
          
        #For orthogonal (k*N, M) M>N
        #z_zero = torch.cat([n.fc1.weight for n in self.targets.values()],1)
        #z_one = orth(z_zero.detach().cpu().numpy())
        #z_one = torch.cuda.FloatTensor(z_one).reshape(8000,784,10)
        #for c in range(n_classes):
         #   self.targets[f'{c}'].fc1.weight=torch.nn.Parameter(z_one[:,:,c])
                
    def activate_class(self, class_):
        self.activated_target = self.targets[f'{class_}']
        self.activated_predictor = self.predictors[f'{class_}']
        self.activated_predictor_next = self.predictors_next[f'{class_}']

        
    def get_optimizer(self, class_i):
        return self.optimizers[f"{class_i}"], self.optimizers_next[f"{class_i}"]
                
        
    def unactivated_predictors(self, features):
        for predictor in self.predictors:
            if self.predictors[predictor] != self.activated_predictor:
                features = self.predictors[predictor](features)
        return features
    
    
    def forward(self, next_obs):
        target = self.activated_target(next_obs)
        predict = self.activated_predictor(next_obs)
        predict_next = self.activated_predictor_next(next_obs)
        return target, predict, predict_next             
            
        
    def predict(self, next_obs):
        target_predicts = []
        predictor_predicts = []
        for predictor in self.predictors:
            self.activated_target = self.targets[predictor]
            self.activated_predictor = self.predictors[predictor]
            self.activated_predictor_next = self.predictors_next[predictor]
            target = self.activated_target(next_obs)
            predict = self.activated_predictor(next_obs)
            predict_next = self.activated_predictor_next(next_obs)
            
            target_predicts.append([target, predict])
            predictor_predicts.append([predict, predict_next])
            
        return target_predicts, predictor_predicts 
    
        
    def to(self, device):
        super(RNDModel, self).to(device)
        # Move all predictor networks to the same device
        if torch.cuda.is_available():
            for predictor in self.predictors:
                self.targets[predictor].to(device)
                self.predictors[predictor].to(device)
                self.predictors_next[predictor].to(device)
```

```python
def get_few_shot_mnist(data_loader, shot=10, classes=10):
    few_shot_dataset = []
    class_counter = Counter()
    for batch_i, (x, y) in enumerate(data_loader):
        if y<classes:
            if class_counter[y.item()] < shot:
                class_counter[y.item()] += 1
                few_shot_dataset.append((x, y))
            if all([x == shot for x in class_counter.values()]):
                break
    return few_shot_dataset
```

### PCA, TSNE

```python
def tsne_plot(x, y, n_sne = 60000):
    feat_cols = [ 'pixel'+str(i) for i in range(x.shape[1]) ]
    df = pd.DataFrame(x,columns=feat_cols)
    df['label'] = y
    df['label'] = df['label'].apply(lambda i: str(i))

    print ('Size of the dataframe:', df.shape)
    rndperm = np.random.permutation(df.shape[0])
    time_start = time.time()
    tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
    tsne_results = tsne.fit_transform(df.loc[rndperm[:n_sne],feat_cols].values)

    df_tsne = df.loc[rndperm[:n_sne],:].copy()
    df_tsne['x-tsne'] = tsne_results[:,0]
    df_tsne['y-tsne'] = tsne_results[:,1]

    chart = ggplot( df_tsne, aes(x='x-tsne', y='y-tsne', color='label') ) \
            + geom_point(size=10,alpha=0.8) \
            + ggtitle("tSNE dimensions colored by digit")

    return chart
```

```python
def pca_plot(x, y):
    feat_cols = [ 'pixel'+str(i) for i in range(x.shape[1]) ]
    df = pd.DataFrame(x,columns=feat_cols)
    df['label'] = y
    df['label'] = df['label'].apply(lambda i: str(i))

    print ('Size of the dataframe:', df.shape)
    rndperm = np.random.permutation(df.shape[0])
    pca = PCA(n_components=3)
    pca_result = pca.fit_transform(df[feat_cols].values)

    df['pca-one'] = pca_result[:,0]
    df['pca-two'] = pca_result[:,1] 
    df['pca-three'] = pca_result[:,2]
    chart = ggplot( df.loc[rndperm[:60000],:], aes(x='pca-one', y='pca-two', color='label') ) \
        + geom_point(size=100,alpha=1) \
        + ggtitle("First and Second Principal Components")

    return chart
```
