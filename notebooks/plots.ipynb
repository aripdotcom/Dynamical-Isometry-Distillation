```python
sns.set_style('white')

data = pd.read_csv('/home/igor/deployments/Dynamical-Isometry-Distillation/results/2019-05-22_15:21.csv')
data_train = data[data['split'] == 'train']
data_train = data_train[data_train['trial'] == 1]

data_train = data_train.rename(columns={'loss/val': 'Loss'})

flatui = ["#3366cc", # Student 0 
          "#0066cc", # Student 1
          "#0099ff", # Student 2 
          "#3399ff", # Student 3
          "#99ccff", # Sutdent 4
          "#cc99ff", # Student 5
          "#ccccff", # Student 6
          "#6699ff", # Student 7
          "#9999ff", # Student 8
          "#6699ff", # Student 9
          "#cc0066"] # Teacher
sns.set_palette(flatui)

hue_order  = ['Student 0',
             'Student 1',
             'Student 2',
             'Student 3',
             'Student 4',
             'Student 5',
             'Student 6',
             'Student 7',
             'Student 8',
             'Student 9',
              'Teacher']

last_trial = data_train.trial.unique()[-1]
data_train['Epoch'] = data_train.epoch.apply(lambda x: int(x))
data_train['Network'] = data_train.predictor
data_train['Sample'] = data_train['epoch'] * 50 + data_train['sample']
for i_trial in data_train.trial.unique():
    trial_data = data_train[data_train['trial'] == i_trial]
    figure(num=0, figsize=(10, 5), dpi=200, facecolor='w', edgecolor='k')        
    if i_trial == last_trial:
        sns.lineplot(data=trial_data, x='Sample', y='Loss', hue='Network', legend='full', palette=sns.color_palette(), hue_order=hue_order)
    else:
        sns.lineplot(data=trial_data, x='Sample', y='Loss', hue='Network', legend=None, palette=sns.color_palette(), hue_order=hue_order)
```
